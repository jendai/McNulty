### Data exploration

```python
#transfer data into pivot tables

def reshapeSeries(s, header):
    out, h, count = list(), list(), list()
    for i, j in s.iteritems():
        tmp = list(i)
        out.append(tmp[0])
        h.append(tmp[1])
        count.append(j)
    tmp = pd.DataFrame()
    tmp['out'] = out
    tmp[header] = h
    tmp['count'] = count

    return tmp.pivot('out', header, 'count')
```

```python
#output variable, y
mpl.rc("savefig", dpi=100)
#y - has the client subscribed a term deposit? (binary: "yes","no")
df['y'].value_counts().plot(kind='barh', colormap='cool')
#majority of the clients said 'no' to short term deposits
df['y'].value_counts()/len(df)
```

```python
def comparisonPlot(feature1, feature2):
    mpl.rc('savefig', dpi=120)
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(17,4))
    df[feature2].value_counts().plot(kind='barh', ax=axes[0], colormap='cool')
    groups = df.groupby(feature1)[feature2].value_counts()
    groups = reshapeSeries(groups, feature2)
    (groups.ix['yes']/groups.sum()).plot(kind='barh', ax=axes[1], colormap='cool')
```

* Tranform data for modeling

```python
df = df.replace('yes', int(1))
df = df.replace('no', int(0))
df = df.replace('y', int(1))
df = df.replace('n', int(0))
df = df.replace('unknown', np.nan)
df['pdays'] = df.replace(999, int(0))
```

```python
from sklearn import preprocessing

#normalize all numeric columns
names = ['age', 'campaign', 'pdays', 'emp_var_rate', 'cons_price_idx', 'cons_conf_idx', 
         'euribor3m', 'nr_employed']

for i in range(len(names)):
    df[names[i]] = df[names[i]].astype(np.float64)
    df[names[i]] = preprocessing.scale(df[names[i]])

#replace all columns with Nan values to column mode
names = df.columns.values

for i in range(len(names)):
    m = df[names[i]].mode()
    df[names[i]] = df[names[i]].replace(np.nan, m[0])
```

### turn categorical variable into dummies

```python
#drop duration from data becasue it cannot be used to fit the model
df = df.drop('duration', 1)
cols = ['job', 'marital', 'education', 'contact', 'contact_month', 'day_of_week', 'poutcome']
df = df.join(pd.get_dummies(df[cols]))
df = df.drop(cols, 1)
```

### Use recursive feature elimination to find the best features for this model

```python
from sklearn.svm import SVC
from sklearn.cross_validation import KFold, StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.cross_validation import train_test_split
```

```python
#split into train and test dataset
y = df['y']
x = df.drop('y', 1)
```

```python
x.columns.values
```

```python
lr = LogisticRegression()
rfecv = RFECV(estimator=lr, step=1, cv=10, scoring='accuracy')
rfecv.fit(x, y)
print("Optimal number of features : %d" % rfecv.n_features_)
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()
```

```python
import operator
import pprint
pp = pprint.PrettyPrinter(indent=4)
features = x.columns.values
best = {}
for i in range(len(features)):
    best[features[i]] = rfecv.ranking_[i]
best = sorted(best.items(), key=operator.itemgetter(1))
pp.pprint (best)
```

```python
from sklearn.feature_selection import SelectPercentile, f_classif
```

```python
plt.figure(1)
plt.clf()

X_indices = np.arange(x.shape[-1])
selector = SelectPercentile(f_classif, percentile=10)
selector.fit(x, y)
scores = -np.log10(selector.pvalues_)
#scores /= scores.max()
print scores
plt.bar(X_indices - .45, scores, width=.2,
        label=r'Univariate score ($-Log(p_{value})$)', color='g')
```

```python
#clf = SVC(kernel='linear')
lr.fit(x, y)
mpl.rc('savefig', dpi=120)
svm_weights = (lr.coef_ ** 2).sum(axis=0)
svm_weights /= svm_weights.max()

plt.bar(X_indices - .25, svm_weights, width=.2, label='LogisticRegression weight', color='r')

clf_selected = SVC(kernel='linear')
lr.fit(selector.transform(x), y)

svm_weights_selected = (lr.coef_ ** 2).sum(axis=0)
svm_weights_selected /= svm_weights_selected.max()

plt.bar(X_indices[selector.get_support()] - .05, svm_weights_selected,
        width=.2, label='LogiticRegression weights after selection', color='b')


plt.title("Comparing feature selection")
plt.xlabel('Feature number')
plt.yticks(())
plt.axis('tight')
plt.legend(loc='upper right')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc, classification_report
```

```python
#split into train and test size
x_train, x_test, y_train, y_test = train_test_split(new_x, y, test_size=0.25)
```

```python
lr = LogisticRegression(tol=0.0000005)
lr = lr.fit(x_train, y_train)
lr_pred_prob = lr.predict_proba(x_test)
lr_pred = lr.predict(x_test)
lr_fpr, lr_tpr, thresholds = roc_curve(y_test, lr_pred_prob[:,1])
lr_auc = auc(lr_fpr, lr_tpr)
lr_class = classification_report(y_test, lr_pred)
print lr_class
print
print lr_auc
plt.plot(lr_fpr, lr_tpr)
```
